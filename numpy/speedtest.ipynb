```python
#using numpy list
import numpy as np
import time

size = 1_000_000
list1 = list(range(size))
list2 =  list(range(size))
start = time.time()
add = [x+y for x,y in zip(list1, list2)]
end = time.time()
print(end - start)
```

```python
#using numpy arrays
import numpy as np
import time

size = 1_000_000
l1 = np.array(list(range(size)))
l2 =  np.array(list(range(size)))

start = time.time()
add = l1 + l2

end = time.time()
print(add)
print(end - start)
```
